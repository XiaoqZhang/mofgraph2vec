```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ori_emb_path = "../experiments/workflow/wandb/offline-run-20230309_132802-2ysz2d4z/tmp/embedding_dv.csv"
task_emb_path = "../experiments/workflow/wandb/offline-run-20230309_132802-2ysz2d4z/tmp/embedding-outputs.pbe.bandgap.csv"
```

```python
ori_emb = pd.read_csv(ori_emb_path)
task_emb = pd.read_csv(task_emb_path)
```

```python
ori_emb[ori_emb["type"] == "qmof-0000295"].values
```

```python
task_emb[task_emb["type"] == "qmof-0000295"]
```

```python
a = ori_emb[ori_emb["type"] == "qmof-0000295"]._get_numeric_data().values
b = task_emb[task_emb["type"] == "qmof-0000295"]._get_numeric_data().values
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
for name in ori_emb["type"]:
    print(name)
```

```python
cosine_similarity(a, b)
```

```python
sim = []
for name in ori_emb["type"]:
    a = ori_emb[ori_emb["type"] == name]._get_numeric_data().values
    b = task_emb[task_emb["type"] == name]._get_numeric_data().values
    s = cosine_similarity(a, b)
    sim.append(s)
```

```python
np.array(sim).flatten()
```

```python
plt.hist(np.array(sim).flatten(), bins=100)
```
